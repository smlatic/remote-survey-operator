```python
import pandas as pd
import openpyxl
import dash
import plotly.express as px
from dash import dcc, html
from dash.dependencies import Input, Output

# Read the Excel file
xlsx = pd.ExcelFile('marine_survey_analysis.xlsx')
sheets = xlsx.sheet_names  # Get sheet names

# Create a DataFrame dictionary to hold data of each sheet
dfs = {sheet: pd.read_excel(xlsx, sheet) for sheet in sheets}

# Start building the app
app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id="dropdown",
        options=[{"label": sheet, "value": sheet} for sheet in sheets],
        value=sheets[0]
    ),
    dcc.Graph(id="graph"),
])

@app.callback(
    Output("graph", "figure"), 
    [Input("dropdown", "value")]
)
def update_graph(selected_sheet):
    df = dfs[selected_sheet]
    return px.line(df)  # Modify as per your needs

if __name__ == "__main__":
    app.run_server(debug=True)
```
